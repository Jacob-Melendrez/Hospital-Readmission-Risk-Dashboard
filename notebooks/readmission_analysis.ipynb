```python
df = df.replace("?", pd.NA)
```

```python
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
df["is_readmitted_30"] = df["readmitted"].apply(lambda x: 1 
    if x == "<30" else 0)
df["is_readmitted_30"].value_counts()
```

```python
cols_to_drop = ["weight", "payer_code", "medical_specialty"]

df = df.drop(columns=cols_to_drop, errors="ignore")
```

```python
df.columns.tolist()
```

```python
analysis_cols = [
    "encounter_id",
    "patient_nbr",
    "race",
    "gender",
    "age",
    "admission_type_id",
    "discharge_disposition_id",
    "time_in_hospital",
    "num_lab_procedures",
    "num_procedures",
    "num_medications",
    "number_outpatient",
    "number_emergency",
    "number_inpatient",
    "diag_1",
    "A1Cresult",
    "insulin",
    "diabetesMed",
    "readmitted",
    "is_readmitted_30"
]

analysis_df = df[analysis_cols].copy()
print(analysis_df.shape)
analysis_df.head()
```

```python
# Overall 30 Day readmission rate.
```

```python
analysis_df["is_readmitted_30"].mean()
```

```python
#Readmission counts
```

```python
analysis_df["is_readmitted_30"].value_counts()
```

```python
#Average length of stay by readmission. 
```

```python
analysis_df.groupby("is_readmitted_30")["time_in_hospital"].mean()
```

```python
#Average number of medications by readmission. 
```

```python
analysis_df.groupby("is_readmitted_30")["num_medications"].mean()
```

```python
#Average number of emergency visits by readmission. 
analysis_df.groupby("is_readmitted_30")["number_emergency"].mean()
```

```python
#Readmission rate by gender. 
analysis_df.groupby("gender")["is_readmitted_30"].mean()
```

```python
#Readmission by age 
analysis_df.groupby("age")["is_readmitted_30"].mean().sort_index()
```

```python
# Creating a midpoint range for cleaner python analysis. 
age_map = {
    "[0-10)": 5, "[10-20)": 15, "[20-30)": 25, "[30-40)": 35,
    "[40-50)": 45, "[50-60)": 55, "[60-70)": 65, "[70-80)": 75,
    "[80-90)": 85, "[90-100)": 95
}

analysis_df["age_mid"] = analysis_df["age"].map(age_map)
analysis_df[["age", "age_mid"]].head()
```

```python
#Readmission rate by age but cleaner. 
analysis_df.groupby("age_mid")["is_readmitted_30"].mean()
```

```python
# Saving the cleaned file for Tableau. 
analysis_df.to_csv("../data/cleaned_readmission_data.csv", index=False)
```
